```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="smollm:latest",
    temperature=0,
    # other params...
)
```

```python
messages = [
    ("system", "You are a helpful assistant."), 
    ("user", "What is the capital of France?")
    ]
```

```python
result = llm.invoke(messages)

print(result)
```
